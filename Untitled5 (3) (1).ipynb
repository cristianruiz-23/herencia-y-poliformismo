```python
class Persona():
  def _init_(self):
    self.__Nombre=""
    self.__Apellido=""
    self.__Edad=0

  def GetNombre(self):
    return self.__Nombre

  def SetNombre(self, nombre):
    self.__Nombre=nombre
  
  def GetApellido(self):
    return self.__Apellido

  def SetApellido(self, apellido):
    self.__Apellido = apellido

  def GetEdad(self):
    return self.__Edad

  def SetEdad(self, edad):
    self.__Edad=edad


#Alumno
  
class Alumno(Persona):
  def _init_(self):
    self.__Curso=""

    self.__Asignatura=""

  def GetCurso(self):
    return self.__Curso

  def SetCurso(self, curso):
    self.__Curso= curso

  def GetAsignatura(self):
    return self.__Asignatura
  
  def SetAsignatura(self, asignatura):
    self.__Asignatura = asignatura 

  def MostrarAlumno(self):
    print("Alumno: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tCurso:",self.__Curso)
    print("\tMatricula:",self.__Asignatura)
    

alumno=Alumno()
alumno.SetNombre("cristian ")
alumno.SetApellido("ruiz")
alumno.SetEdad("17")
alumno.SetCurso("curso 10")
alumno.SetAsignatura(["sociales","ingles","historia"])
alumno.MostrarAlumno()


#Profesor
class Profesor(Persona):
  def _init_(self):
    self.__Antiguedad=""
    self.__Tutorias=""
    self.__Telefono=""

  def GetAntiguedad(self):
    return self.__Antiguedad

  def SetAntiguedad(self, antiguedad):
    self.__Antiguedad = antiguedad

  def GetTutorias(self):
    return self.__Tutorias

  def SetTutorias(self, tutorias):
    self.__Tutorias = tutorias

  def GetTelefono(self):
    return self.__Telefono

  def SetTelefono(self, telefono):
    self.__Telefono = telefono

  def MostrarProfesor(self):
    print("Profesor: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tAntiguedad:",self.__Antiguedad)
    print("\tTutorias:",self.__Tutorias)
    print("\tTelefono:",self.__Telefono)


profesor = Profesor()
profesor.SetNombre("cristian")
profesor.SetApellido("sarmiento")
profesor.SetEdad(25)
profesor.SetAntiguedad(5)
profesor.SetTutorias([["Horarios: ","lunes:"+"6-12 pm","miercoles:"+"6-12 pm"]])
profesor.SetTelefono("3197051541")
profesor.MostrarProfesor()


#Investigador:
class Investigador(Persona):
  def _init_(self):
    self.__Atributos=""
    self.__Especialidad=""
    self.__Años=""

  def GetAtributos(self):
    return self.__Atributos

  def SetAtributos(self, atributos):
    self.__Atributos = atributos

  def GetEspecialidad(self):
    return self.__Especialidad

  def SetEspecialidad(self, especialidad):
    self.__Especialidad = especialidad

  def GetAños(self):
    return self.__Años

  def SetAños(self, años):
    self.__Años = años

  def MostrarInvestigador(self):
    print("Investigador: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tAtributos:",self.__Atributos)
    print("\tEspecialidad:",self.__Especialidad)
    print("\tAños:",self.__Años)


investigador = Investigador()
investigador.SetNombre("cristian ")
investigador.SetApellido("sarmiento")
investigador.SetEdad(30)
investigador.SetAtributos("arquitecto")
investigador.SetEspecialidad("Investigador")
investigador.SetAños(5)
investigador.MostrarInvestigador()

#Profesor_Universidad

class Profesor_Universidad(Persona):
  def _init_(self):
    self.__Atributos=""
    self.__Universidad=""
    self.__Departamento=""

  def GetAtributos(self):
    return self.__Atributos

  def SetAtributos(self, atributos):
    self.__Atributos = atributos

  def GetUniversidad(self):
    return self.__Universidad

  def SetUniversidad(self, universidad):
    self.__Universidad = universidad

  def GetDepartamento(self):
    return self.__Departamento

  def SetDepartamento(self, departamento):
    self.__Departamento = departamento

  def MostrarProfesor_Universidad(self):
    print("Profesor de Universidad: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tAtributos:",self.__Atributos)
    print("\tUniversidad:",self.__Universidad)
    print("\tDepartamento:",self.__Departamento)

    class Profesor_Universitario(Persona and Investigador):
       def _init_(self):
          self.__Universidad=""
    self.__Departamento=""

  def GetUniversidad(self):
    return self.__Universidad

  def SetUniversidad(self, universidad):
    self.__Universidad = universidad

  def GetDepartamento(self):
    return self.__Departamento

  def SetDepartamento(self, departamento):
    self.__Departamento = departamento

  def MostrarProfesor_Universidad(self):
    print("Profesor Universitario: ")
    print("\tNombre:",self.GetNombre())
    print("\tApellido:",self.GetApellido())
    print("\tEdad:",self.GetEdad())
    print("\tAntiguedad",self.GetAntiguedad())
    print("\tTutorias:",self.GetTutorias())
    print("\tTelefono:",self.GetTelefono())
    print("\tEspecialidad",self.GetEspecialidad())
    print("\tAños:",self.GetAños())
    print("\tUniversidad:",self.__Universidad)
    print("\tDepartamento:",self.__Departamento)


profe_universidad = profesor_universidad()
profe_universidad.SetNombre("CRISTIAN ALIRIO")
profe_universidad.SetApellido("RUIZ SARMIENTO")
profe_universidad.SetEdad(30)
profe_universidad.SetAntiguedad(7)
profe_universidad.SetTutorias([["Horarios: ","LUNES:"+"6-12 pm","Miercoles:"+"6-12 pm"]])
profe_universidad.SetTelefono("3197051541")
profe_universidad.SetEspecialidad("INGENIERIA SOFTWARE")
profe_universidad.SetAños("5")
profe_universidad.SetUniversidad("FET- fundacion escuela tecnlogica de neiva")
profe_universidad.SetDepartamento("NEIVA-HUILA")
profe_universidad.MostrarProfesor_Universidad() 
```

```python
class Animal():
  def init(self):
    self.__Color=""
    self.__Peso=""
    self.__Comer=""
    self.__Moverse=""
   

  def GetColor(self):
    return self.__Color

  def SetColor(self, color):
    self.__Color= color
  
  def GetPeso(self):
    return self.__Peso

  def SetPeso(self, peso):
    self.__Peso = peso

  def GetComer(self):
    return self.__Comer

  def SetComer(self, comer):
    self.__Comer = comer

  def GetMoverse(self):
    return self.__Moverse

  def SetMoverse(self, moverse):
    self.__Moverse = moverse


class Ave(Animal):
  def init(self):
    self.__Tipo=""
    self.__Volar=""
    

  def GetTipo(self):
    return self.__Tipo

  def SetTipo(self, tipo):
    self.__Tipo=tipo
  
  def GetVolar(self):
    return self.__Volar

  def SetVolar(self, volar):
    self.__Volar = volar

  def MostrarAve(self):
    print("ave: ")
    print("\tColor:",self.GetColor())
    print("\tPeso:",self.GetPeso())
    print("\tComer:",self.GetComer())
    print("\tMoverse:",self.GetMoverse())
    print("\tTipo:",self.GetTipo())
    print("\tVolar:",self.GetVolar())

ave=Ave()
ave.SetColor("verde oscurito")
ave.SetPeso("4 kg")
ave.SetComer("semillas")
ave.SetMoverse("veloz")
ave.SetTipo("guacamaya")
ave.SetVolar("rapidamente")
ave.MostrarAve()

class Perro(Animal):
  def init(self):
    self.__Raza=""
    self.__Ladra=""
    

  def GetRaza(self):
    return self.__Raza

  def SetRaza(self, raza):
    self.__Raza=raza
  
  def GetLadra(self):
    return self.__Ladra

  def SetLadra(self, ladra):
    self.__Ladra = ladra

  def MostrarPerro(self):
    print("perro: ")
    print("\tColor:",self.GetColor())
    print("\tPeso:",self.GetPeso())
    print("\tComer:",self.GetComer())
    print("\tMoverse:",self.GetMoverse())
    print("\tRaza:",self.GetRaza())
    print("\tLadra:",self.GetLadra())

perro=Perro()
perro.SetColor("negro")
perro.SetPeso("14 kg")
perro.SetComer("purina")
perro.SetMoverse("en 4 patas")
perro.SetRaza("labrador")
perro.SetLadra("feroz y mueve su cola")
perro.MostrarPerro()    


  
```
